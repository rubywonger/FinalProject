```python
import random

class mastermind():
    
    def __init__(self):
        self.secret_code = []
        self.allowed_characters = ['1', '2', '3', '4']
        self.game = False
        self.attempt_check = None
        
    def generate_code(self):
        self.secret_code.clear()
        for i in range(4):
            self.secret_code.append(random.choice(self.allowed_characters))
        return self.secret_code
    
    def wrong_length(self, attempt):
        return len(attempt) != 4
    
    def wrong_characters(self, attempt):
        for n in attempt:
            return n not in self.allowed_characters
    
    def get_attempt(self, attempt):
        if self.wrong_characters(attempt) == True:
            print('Your guess must only be numbers 1-4, try again!')
            return False
        elif self.wrong_length(attempt) == True:
            print("Your guess must be exactly four digits long, try again!")
            return False
        if self.wrong_length(attempt) == False and self.wrong_characters(attempt) == False:
            return True
    
    def check_numbers(self, attempt, answer):
        number_check = 0
        order_check = 0
        empty = 0
        list_attempt = [int(l) for l in attempt]
        if list_attempt == answer:
            print("Congratulations! You guessed the secret code!")
            self.game = False
        elif list_attempt != answer:
            for i in range(4):
                if attempt[i] == self.secret_code[i]:
                    order_check += 1
                elif list_attempt[i] != self.secret_code[i]:
                    if attempt[i] in self.secret_code:
                        number_check += 1
                    if attempt[i] not in self.secret_code:
                        empty += 1
            print("Your guess has", number_check, "in the wrong order,",
                    order_check, "in the correct order, and",
                    empty, "not in the secret code!")
    
    def continue_game_condition(self, attempt, attempts, answer):
        list_attempt = [int(l) for l in attempt]
        if attempts < 10:
            if list_attempt != answer:
                print ("You have", 10 - attempts, "guesses left!")
        elif attempts == 10:
            print("Sorry, you ran out of attempts!"
                   "\nThe secret code was", answer)
            self.game = False
        
    def play_game(self):
        acceptable_input = ["YES", "SURE", "ABSOLUTELY I DO", "YEA"]
        print("Lets play a game!"
        "\nHow to play: You get 10 tries to guess the 4-digit secret code created by the Code Maker."
        "\nThe secret code will be a combination of numbers 1-4."
        "\nThe codemaker will tell you how many numbers of your attempt are:"
        "\ncorrect and in the right order, correct and in the wrong order, and not even in the secret code!")
        if self.game == False:
            start_game = input("Want to play? Enter Yes, Sure, or Absolutely I do : ").upper()
            if start_game in acceptable_input:
                self.generate_code()
                attempts = 0
                self.game = True
            if start_game == "NO":
                return print("Bye Felicia!")
            elif start_game not in acceptable_input and start_game != "NO":
                print("Enter Yes, Sure, Absolutely I do, ...or no")
        while self.game == True:
            attempt = input("Enter guess: ")
            answer = [int(i) for i in self.secret_code]
            if self.get_attempt(attempt) == False:
                print("Try again!")
            elif self.get_attempt(attempt) == True:
                attempts += 1
                self.continue_game_condition(attempt,attempts,answer)
                self.check_numbers(attempt,answer)
```

```python
import unittest

class TestMastermind(unittest.TestCase):
        
    def test_wrong_length(self):
        cases = [(1, 2, 3, 5, 4),
                 (1, 2, 4)
                ]
        for case in cases:
            self.assertTrue(mastermind().wrong_length(case))
            
    def test_wrong_characters(self):
        cases = [('q', 'w', 'e', 'r'),
                 (3, 4, 5, 6),
                ]
        for case in cases:
            self.assertIs(mastermind().wrong_characters(case), True)
            
    def test_get_attempt(self):
        cases = [(1, 2, 3, 4)
                ]
        for case in cases:
            self.assertFalse(mastermind().get_attempt(case))
    
unittest.main(argv=[''], verbosity=2, exit=False)
```
